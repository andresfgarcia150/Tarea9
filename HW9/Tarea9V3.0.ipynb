```python
# Reads the input file
obs_data = loadtxt('traces.dat')
x1_obs = obs_data[:,0]
y1_obs = obs_data[:,1]
z1_obs = obs_data[:,2]
x2_obs = obs_data[:,3]
y2_obs = obs_data[:,4]
z2_obs = obs_data[:,5]
```

```python
def likelihood(x_obs, x_pred, y_obs, y_pred):
    chi_2 = sum((x_pred - x_obs)**2 + (y_pred - y_obs)**2)
    return (-chi_2)

def predX(x0, z0, x1, z1, z2_obs):
    m = (x1 - x0)/(z1 - z0)
    b = x1 - m * z1
    return (m * z2_obs + b)

def predY(y0, z0, y1, z1, z2_obs):
    m = (y1 - y0)/(z1 - z0)
    b = y1 - m * z1
    return (m * z2_obs + b)
```

```python
x_walk = empty((0))
y_walk = empty((0))
z_walk = empty((0))

x_walk = append(x_walk, 133)
y_walk = append(y_walk, 89)
z_walk = append(z_walk, -169)
```

```python
n_iter = 50000

for i in range(n_iter):
    x_prima = np.random.normal(x_walk[i],0.1)
    y_prima = np.random.normal(y_walk[i],0.1)
    z_prima = np.random.normal(z_walk[i],20.0)
    while ((z_prima > 0.0) | (z_prima < -10000.0)):
        z_prima = np.random.normal(z_walk[i],20.0)
    
    y2_init = predY(y_walk[i], z_walk[i], y1_obs, z1_obs, z2_obs)
    y2_prima = predY(y_prima, z_prima, y1_obs, z1_obs, z2_obs)
    x2_init = predX(x_walk[i], z_walk[i], x1_obs, z1_obs, z2_obs)
    x2_prima = predX(x_prima, z_prima, x1_obs, z1_obs, z2_obs)
    
    alpha = likelihood(x2_obs, x2_prima, y2_obs, y2_prima)/likelihood(x2_obs, x2_init, y2_obs, y2_init)
    
    if (alpha <= 1.0):
        x_walk = append(x_walk, x_prima)
        y_walk = append(y_walk, y_prima)
        z_walk = append(z_walk, z_prima)
    else:
        beta = random.random()
        if (beta <= exp(-alpha)):
            x_walk = append(x_walk, x_prima)
            y_walk = append(y_walk, y_prima)
            z_walk = append(z_walk, z_prima)
        else:
            x_walk = append(x_walk, x_walk[i])
            y_walk = append(y_walk, y_walk[i])
            z_walk = append(z_walk, z_walk[i])
```
